```python
# Definici√≥n de las clasificaciones del IMC
'''
IMC = BAJO PESO   = < 18.5
IMC = PESO NORMAL = 18.5 <= IMC < 25.0
IMC = SOBREPESO   = 25.0 <= IMC < 30.0
IMC = OBESIDAD    = >= 30.0
'''

# DATOS TABULADOS O EN DICCIONARIO (Estructura de Datos: Lista de Diccionarios)
lista_personas = [
    {'id': 1, 'peso': 62, 'altura': 1.65},  # Diccionario
    {'id': 2, 'peso': 72, 'altura': 1.73},  # Diccionario
    {'id': 3, 'peso': 65, 'altura': 1.57},  # Diccionario
    {'id': 4, 'peso': 70, 'altura': 1.68},  # Diccionario
    {'id': 5, 'peso': 95, 'altura': 1.89}   # Diccionario
]

# --- Funciones ---

##  Calcular IMC
def calcular_imc(peso, altura):
    """Calcula el √çndice de Masa Corporal (IMC): PESO / ESTATURA**2"""
    # Se a√±ade un manejo b√°sico de error para evitar la divisi√≥n por cero
    if altura <= 0:
        return 0
    return peso / (altura ** 2)

## üè∑Ô∏è Clasificar IMC
def clasificar_imc(imc):
    """Clasifica el IMC seg√∫n los rangos establecidos."""
    if imc < 18.5:
        return 'Bajo peso'
    elif imc < 25.0:  # Esto es equivalente a 18.5 <= imc < 25.0
        return 'Peso Normal'
    elif imc < 30.0:  # Esto es equivalente a 25.0 <= imc < 30.0
        return 'Sobrepeso'
    else:  # Esto cubre IMC >= 30.0
        return 'Obesidad'

# --- Ciclo de Ejecuci√≥n (Estructura Iterativa) ---

##  An√°lisis de IMC para cada persona
print("--- Resultado del An√°lisis de IMC ---")
print("{:<10} {:>5} {:>7} {:>6} {:<15}".format("ID", "Peso", "Altura", "IMC", "Clasificaci√≥n"))
print("-" * 45)

# Recorrer la lista de personas (ciclo iterativo 'for')
for persona in lista_personas:
    # 1. Extraer datos (acceder a los valores del diccionario)
    id_persona = persona['id']
    peso = persona['peso']
    altura = persona['altura']

    # 2. Calcular el IMC
    imc = calcular_imc(peso, altura)

    # 3. Clasificar el IMC
    clasificacion = clasificar_imc(imc)

    # 4. Imprimir el resultado (Formateado para una mejor presentaci√≥n)
    # Se usa f-string y formato de texto para alinear las columnas
    print(f"{id_persona:<10} {peso:>5} {altura:>7.2f} {imc:>6.2f} {clasificacion:<15}")

print("-" * 45)
```
