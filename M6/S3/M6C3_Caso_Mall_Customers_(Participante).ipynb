# Caso Mall Customers

Se cuenta con un dataset de clientes de un importante mall de la ciudad. Los datos son los siguientes:

- CustomerID: correlativo identificador del cliente
- Genre: sexo del cliente
- Age: edad del cliente
- Annual Income: sueldo anual en miles de dolares
- Spending Score: score de gasto calculado para el cliente, mientras mayor más gasta

Se busca realizar una caracterización de los clientes, para lo cual se le solicita que realice una **Clusterización Jerárquica** considerando solamente las variables **Annual Income** y **Spending Score**. Se pide lo siguiente:

- Realice una clusterización K-Means
- Aplique método del codo para determinar la cantidad de clusters
- Aplique método de la silueta
- Agregue la clusterización al dataset

---

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Carga de datos

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset.describe()
```

### Formulación del Modelo

```python
# Usaremos solo las columnas de interés: Annual Income (k$) y Spending Score (1-100)
X = dataset[["Annual Income (k$)", "Spending Score (1-100)"]].values
```

### Escalamiento

```python
from sklearn.preprocessing import StandardScaler

# Escalamiento de las variables (recomendado para clustering por distancia)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

```python
# Opcional: comparar distribución antes y después del escalamiento
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(X[:, 0], X[:, 1])
axes[0].set_title('Sin escalar')
axes[0].set_xlabel('Annual Income')
axes[0].set_ylabel('Spending Score')

axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1])
axes[1].set_title('Escalado (StandardScaler)')
axes[1].set_xlabel('Annual Income (scaled)')
axes[1].set_ylabel('Spending Score (scaled)')

plt.tight_layout()
plt.show()
```

```python
# Para el resto del ejercicio utilizaremos X_scaled como entrada al algoritmo
X_for_clust = X_scaled
```

### Clusterización K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Definimos el rango de k a evaluar
range_n_clusters = range(2, 11)
inertias = []
silhouettes = []

for k in range_n_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_for_clust)
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(X_for_clust, kmeans.labels_))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(list(range_n_clusters), inertias, marker='o')
plt.title('Método del codo (inercia) - K-Means')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')

plt.subplot(1, 2, 2)
plt.plot(list(range_n_clusters), silhouettes, marker='o', color='orange')
plt.title('Coeficiente de silueta promedio - K-Means')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Silueta')

plt.tight_layout()
plt.show()

for k, s in zip(range_n_clusters, silhouettes):
    print(f'k = {k}, silueta promedio = {s:.3f}')
```

```python
# Seleccionamos el mejor k según el coeficiente de silueta
print("Valores de silueta:", silhouettes)
best_k = list(range_n_clusters)[np.argmax(silhouettes)]
print(f"Mejor k según silueta: {best_k}")

# También se podría elegir k visualmente a partir del codo.
# Aquí usaremos best_k como número final de clusters.
n_clusters_final = best_k

kmeans_final = KMeans(n_clusters=n_clusters_final, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_for_clust)

np.unique(cluster_labels, return_counts=True)
```

```python
# Guardamos los labels de K-Means para usarlos en gráficos y análisis posteriores
labels = cluster_labels
np.unique(labels, return_counts=True)
```

```python
# Guardamos los labels para usarlos en gráficos y análisis posteriores
labels = cluster_labels
```

```python
# Resumen rápido de cuántos clientes hay en cada cluster
cluster_ids, counts = np.unique(labels, return_counts=True)
for cid, cnt in zip(cluster_ids, counts):
    print(f"Cluster {cid}: {cnt} clientes")
```

### Grafique los resultados

```python
# Gráfico de dispersión de Annual Income vs Spending Score coloreado por cluster (K-Means)
plt.figure(figsize=(8, 6))

scatter = plt.scatter(
    X[:, 0], X[:, 1],
    c=labels,
    cmap='tab10',
    s=50,
    edgecolor='k'
)

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Clusters de clientes (K-Means)')
plt.legend(*scatter.legend_elements(), title='Cluster')
plt.show()
```

```python
# Boxplots para entender el perfil de cada cluster en las dos variables
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for cid in np.unique(labels):
    axes[0].boxplot(
        [X[labels == cid, 0]],
        positions=[cid],
        widths=0.6
    )
axes[0].set_xticks(cluster_ids)
axes[0].set_xlabel('Cluster')
axes[0].set_ylabel('Annual Income')
axes[0].set_title('Distribución de Annual Income por cluster')

for cid in np.unique(labels):
    axes[1].boxplot(
        [X[labels == cid, 1]],
        positions=[cid],
        widths=0.6
    )
axes[1].set_xticks(cluster_ids)
axes[1].set_xlabel('Cluster')
axes[1].set_ylabel('Spending Score')
axes[1].set_title('Distribución de Spending Score por cluster')

plt.tight_layout()
plt.show()
```

### Modifique dataset original con resultados de la clusterización

```python
# Agregar la etiqueta de cluster al dataset original
# (sumamos 1 para que los clusters vayan de 1..k en lugar de 0..k-1)
dataset['Cluster'] = labels + 1

dataset.head()
```

```python
# Resumen de cada cluster: cantidad de clientes y promedios de las principales variables
cluster_summary = dataset.groupby('Cluster').agg({
    'CustomerID': 'count',
    'Age': 'mean',
    'Annual Income (k$)': 'mean',
    'Spending Score (1-100)': 'mean'
}).rename(columns={'CustomerID': 'NumClientes'})

cluster_summary
```

### 5. Comparación y análisis

#### Determinación del número de clusters
A partir de las métricas de evaluación aplicadas, se aprecia claramente la formación de **5 clusters óptimos**:
* **Método del Codo (Inercia):** En el gráfico de inercia se observa que el punto de inflexión ("codo") más pronunciado ocurre cuando `k=5`. A partir de este valor, la inercia disminuye de forma marginal.
* **Coeficiente de Silueta:** Esta métrica confirma matemáticamente la observación anterior. El valor más alto de la silueta promedio se obtiene con **`k=5` (0.555)**, indicando que es la configuración donde los datos están mejor cohesionados dentro de su grupo y mejor separados del resto.

#### Perfil y características de los Clusters
Analizando el gráfico de dispersión y los promedios de cada grupo en cuanto a **ingreso anual** y **spending score**, podemos definir los siguientes perfiles:

* **Cluster 1 (Promedio / Convencional):** Es el grupo mayoritario (81 clientes). Poseen un ingreso anual medio (aprox. $55k) y un score de gasto medio (49.5). Su edad promedio es de 42.7 años.
* **Cluster 2 (Alto Valor / VIP):** Grupo de 39 clientes. Se caracterizan por tener ingresos altos ($86.5k) y un score de gasto muy alto (82.1). Son relativamente jóvenes (32.6 años promedio). **Representan los clientes de mayor valor para el mall.**
* **Cluster 3 (Impulsivos / Jóvenes gastadores):** Conformado por 22 clientes con ingresos bajos ($25.7k) pero un nivel de gasto altísimo (79.3). Es el grupo más joven (25.2 años), lo que sugiere una alta proporción de su ingreso destinada al consumo.
* **Cluster 4 (Conservadores / Ahorradores):** Son 35 clientes con **ingresos muy altos ($88.2k)**, pero con el **gasto más bajo** del mall (score 17.1). Tienen alto poder adquisitivo pero son muy cuidadosos con sus compras.
* **Cluster 5 (Económicos / Prudentes):** Grupo de 23 clientes con ingresos bajos ($26.3k) y gasto bajo (score 20.9). Es el segmento de mayor edad (45.2 años promedio).

#### Estrategias de Marketing y Fidelización
El mall puede utilizar esta segmentación para dirigir esfuerzos de forma personalizada:
1. **Cluster 2 (VIP):** Implementar programas de lealtad premium, ventas privadas y beneficios exclusivos (valet parking, personal shopper) para mantener su alta retención.
2. **Cluster 3 (Impulsivos):** Lanzar campañas por redes sociales con ofertas relámpago, moda de tendencia (fast-fashion) y facilidades de pago, aprovechando su alta propensión al gasto.
3. **Cluster 4 (Ahorradores):** Para incentivar el gasto de este grupo con alto poder adquisitivo, se deben ofrecer productos de alta gama y durabilidad, respaldados por promociones bancarias (ej. meses sin intereses) o artículos que representen una inversión inteligente.
4. **Cluster 1 (Promedio):** Mantener su recurrencia mediante promociones masivas, cupones de descuento y campañas de temporada (Liquidaciones, Black Friday) para intentar elevar gradualmente su ticket promedio.
