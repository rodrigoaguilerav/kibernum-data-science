# Caso Mall Customers

Se cuenta con un dataset de clientes de un importante mall de la ciudad. Los datos son los siguientes:

- CustomerID: correlativo identificador del cliente
- Genre: sexo del cliente
- Age: edad del cliente
- Annual Income: sueldo anual en miles de dolares
- Spending Score: score de gasto calculado para el cliente, mientras mayor más gasta

Se busca realizar una caracterización de los clientes, para lo cual se le solicita que realice una **Clusterización Jerárquica** considerando solamente las variables **Annual Income** y **Spending Score**. Se busca contestar las siguientes preguntas:

- Cuántos clusters se aprecian
- Qué grupo de clientes es representado en cada cluster

---

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Carga de datos

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset.describe()
```

### Formulación del Modelo

```python
# Usaremos solo las columnas de interés: Annual Income (k$) y Spending Score (1-100)
X = dataset[["Annual Income (k$)", "Spending Score (1-100)"]].values
```

### Escalamiento

```python
from sklearn.preprocessing import StandardScaler

# Escalamiento de las variables (recomendado para clustering por distancia)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

```python
# Opcional: comparar distribución antes y después del escalamiento
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(X[:, 0], X[:, 1])
axes[0].set_title('Sin escalar')
axes[0].set_xlabel('Annual Income')
axes[0].set_ylabel('Spending Score')

axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1])
axes[1].set_title('Escalado (StandardScaler)')
axes[1].set_xlabel('Annual Income (scaled)')
axes[1].set_ylabel('Spending Score (scaled)')

plt.tight_layout()
plt.show()
```

```python
# Para el resto del ejercicio utilizaremos X_scaled como entrada al algoritmo
X_for_clust = X_scaled
```

### Clusterización Jerárquica

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Matriz de ligamiento usando método ward (minimiza varianza intra-cluster)
Z = linkage(X_for_clust, method='ward')

plt.figure(figsize=(12, 6))
plt.title('Dendrograma - Clustering Jerárquico (Ward)')
plt.xlabel('Clientes')
plt.ylabel('Distancia Euclídea')

dendrogram(Z, truncate_mode='lastp', p=30, leaf_rotation=90., leaf_font_size=8.)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Método del codo: probamos distintos números de clusters y calculamos la inercia aproximada
range_n_clusters = range(2, 11)
inertias = []
silhouettes = []

for k in range_n_clusters: # 2,3,4,5,6,7,7,9,10
    hc = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
    labels = hc.fit_predict(X_for_clust)

    # Inercia aproximada: suma de distancias cuadráticas al centroide de cada cluster
    inertia_k = 0
    for c in np.unique(labels):
        cluster_points = X_for_clust[labels == c]
        center = cluster_points.mean(axis=0)
        inertia_k += ((cluster_points - center) ** 2).sum()
    inertias.append(inertia_k)

    # Coeficiente de silueta promedio
    silhouettes.append(silhouette_score(X_for_clust, labels))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(list(range_n_clusters), inertias, marker='o')
plt.title('Método del codo (inercia aproximada)')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')

plt.subplot(1, 2, 2)
plt.plot(list(range_n_clusters), silhouettes, marker='o', color='orange')
plt.title('Coeficiente de silueta promedio')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Silueta')

plt.tight_layout()
plt.show()

for k, s in zip(range_n_clusters, silhouettes):
    print(f'k = {k}, silueta promedio = {s:.3f}')
```

```python
from sklearn.cluster import AgglomerativeClustering

# A partir del dendrograma, el método del codo y la silueta,
# elija un número de clusters razonable. Aquí usaremos 5 como ejemplo.
print("Valores de silueta:", silhouettes)
best_k = list(range_n_clusters)[np.argmax(silhouettes)]
print(f"Mejor k según silueta: {best_k}")

n_clusters_final = 5
hc_final = AgglomerativeClustering(n_clusters=n_clusters_final, metric='euclidean', linkage='ward')
cluster_labels = hc_final.fit_predict(X_for_clust)

np.unique(cluster_labels, return_counts=True)
```

```python
# Guardamos los labels para usarlos en gráficos y análisis posteriores
labels = cluster_labels
```

```python
# Resumen rápido de cuántos clientes hay en cada cluster
cluster_ids, counts = np.unique(labels, return_counts=True)
for cid, cnt in zip(cluster_ids, counts):
    print(f"Cluster {cid}: {cnt} clientes")
```

### Grafique los resultados

```python
# Gráfico de dispersión de Annual Income vs Spending Score coloreado por cluster
plt.figure(figsize=(8, 6))

scatter = plt.scatter(
    X[:, 0], X[:, 1],
    c=labels,
    cmap='tab10',
    s=50,
    edgecolor='k'
)

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Clusters de clientes (Clustering Jerárquico)')
plt.legend(*scatter.legend_elements(), title='Cluster')
plt.show()
```

```python
# Boxplots para entender el perfil de cada cluster en las dos variables
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for cid in np.unique(labels):
    axes[0].boxplot(
        [X[labels == cid, 0]],
        positions=[cid],
        widths=0.6
    )
axes[0].set_xticks(cluster_ids)
axes[0].set_xlabel('Cluster')
axes[0].set_ylabel('Annual Income')
axes[0].set_title('Distribución de Annual Income por cluster')

for cid in np.unique(labels):
    axes[1].boxplot(
        [X[labels == cid, 1]],
        positions=[cid],
        widths=0.6
    )
axes[1].set_xticks(cluster_ids)
axes[1].set_xlabel('Cluster')
axes[1].set_ylabel('Spending Score')
axes[1].set_title('Distribución de Spending Score por cluster')

plt.tight_layout()
plt.show()
```

### Modifique dataset original con resultados de la clusterización

```python
# Agregar la etiqueta de cluster al dataset original
# (sumamos 1 para que los clusters vayan de 1..k en lugar de 0..k-1)
dataset['Cluster'] = labels + 1

dataset.head()
```

```python
# Resumen de cada cluster: cantidad de clientes y promedios de las principales variables
cluster_summary = dataset.groupby('Cluster').agg({
    'CustomerID': 'count',
    'Age': 'mean',
    'Annual Income (k$)': 'mean',
    'Spending Score (1-100)': 'mean'
}).rename(columns={'CustomerID': 'NumClientes'})

cluster_summary

```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    cluster_summary['Annual Income (k$)'],
    cluster_summary['Spending Score (1-100)'],
    c=cluster_summary.index,
    cmap='viridis',
    s=200,
    edgecolor='k',
    alpha=0.8
)


for i, cluster_id in enumerate(cluster_summary.index):
    plt.text(
        cluster_summary['Annual Income (k$)'].iloc[i],
        cluster_summary['Spending Score (1-100)'].iloc[i],
        str(cluster_id),
        fontsize=12,
        ha='center',
        va='bottom'
    )

plt.xlabel('Average Annual Income (k$)')
plt.ylabel('Average Spending Score (1-100)')
plt.title('Perfiles de Clientes por Cluster (Promedios)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(np.arange(0, cluster_summary['Annual Income (k$)'].max() + 20, 10))
plt.yticks(np.arange(0, cluster_summary['Spending Score (1-100)'].max() + 20, 10))
plt.colorbar(scatter, label='Cluster ID')
plt.show()
```

## 5. Comparación y análisis

### Conclusiones Clave sobre los Perfiles de Clientes

La clusterización jerárquica de los clientes del mall, utilizando 'Annual Income (k$)' y 'Spending Score (1-100)', ha revelado cinco segmentos de clientes distintivos. Estos segmentos ofrecen una base sólida para estrategias de marketing y fidelización más personalizadas:

1.  **Cluster 2: Clientes de Alto Valor (Alto Ingreso, Alto Gasto)**
    *   Representan el grupo más deseable para el mall. Son clientes con altos ingresos y, crucialmente, con un muy alto Spending Score. Este grupo probablemente está interesado en productos premium, nuevas colecciones y experiencias de compra exclusivas. Las estrategias deberían enfocarse en la retención, programas de lealtad VIP y promociones de alto valor para mantener su elevado nivel de gasto.

2.  **Cluster 4: Compradores Entusiastas (Bajo Ingreso, Alto Gasto)**
    *   Este grupo, aunque con ingresos más bajos, exhibe un Spending Score excepcionalmente alto. Podrían ser clientes jóvenes, sensibles a las tendencias, que priorizan el gasto en el mall o son impulsivos. Las estrategias podrían incluir ofertas, descuentos, financiación asequible y campañas de marketing que enfaticen el valor percibido o la novedad, para incentivar su alta propensión al gasto a pesar de las limitaciones de ingreso.

3.  **Cluster 1: Clientes de Ingresos Altos, pero Frugales (Alto Ingreso, Bajo Gasto)**
    *   Este segmento tiene un alto poder adquisitivo, pero un bajo Spending Score en el mall. Podrían estar gastando su dinero en otro lugar, ser compradores muy selectivos, o simplemente no encuentran lo que buscan en el mall. Es una oportunidad para el mall de atraer una parte de su gasto. Se podrían diseñar campañas enfocadas en productos de lujo, servicios personalizados, o eventos exclusivos que puedan despertar su interés y demostrar el valor del mall.

4.  **Cluster 5: Clientes Frugales (Bajo Ingreso, Bajo Gasto)**
    *   Este grupo presenta tanto bajos ingresos como un bajo Spending Score. Son clientes conscientes de su presupuesto, que posiblemente visitan el mall con poca frecuencia o solo para compras esenciales. Las estrategias deberían enfocarse en atraerlos con promociones de valor, ofertas especiales, productos básicos a precios competitivos o programas de recompensas que hagan que cada compra cuente.

5.  **Cluster 3: Clientes Promedio (Ingreso Medio, Gasto Medio)**
    *   Siendo el grupo más grande, estos clientes representan la base del mall. Sus hábitos de gasto e ingresos son moderados. Las estrategias para este grupo deberían ser amplias, buscando incentivar un ligero aumento en la frecuencia o el volumen de compra, quizás a través de ofertas generales, eventos de temporada o programas de fidelización estándar para convertirlos en clientes más valiosos.

**Utilidad de la Segmentación para el Mall:**

La clusterización jerárquica ha permitido identificar claramente distintos arquetipos de clientes, lo cual es fundamental para:
*   **Marketing Dirigido**: Crear campañas publicitarias y promociones específicas para cada segmento, aumentando la efectividad de los esfuerzos de marketing.
*   **Gestión de Producto**: Adaptar la oferta de productos y servicios a las preferencias y capacidad de gasto de cada grupo.
*   **Optimización de Recursos**: Asignar los recursos de marketing y ventas de manera más eficiente, concentrándose en los segmentos con mayor potencial o en aquellos que necesitan ser activados.
*   **Fidelización**: Desarrollar programas de lealtad y experiencias personalizadas que resuenen con las expectativas de cada tipo de cliente.

En resumen, esta segmentación proporciona una visión valiosa del comportamiento del consumidor, permitiendo al mall implementar estrategias más inteligentes y rentables para cada uno de sus públicos.
