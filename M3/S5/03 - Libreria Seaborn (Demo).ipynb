## Gr√°ficos de Matrices

```python
vuelos = sns.load_dataset('flights')
```

```python
vuelos.head()
```

```python
p = vuelos.pivot_table(index='month', columns='year', values='passengers')
p
```

```python
sns.heatmap(p)
```

```python
sns.heatmap(p, cmap='coolwarm', linecolor='white', linewidth=1,
               annot=True, fmt='d')
```

## Grillas de Gr√°ficos

### 1. Grilla de Facetas

```python
grilla = sns.FacetGrid(data=df, col='day', row='smoker' )
grilla.map(sns.regplot, 'total_bill', 'tip')
```

```python
grilla = sns.FacetGrid(data=df, col='day', row='smoker' )
grilla.map(sns.kdeplot, 'total_bill')
```

```python
grilla = sns.FacetGrid(data=df, col='day', row='sex' )
grilla.map(sns.kdeplot, 'total_bill')
```

```python
grilla = sns.FacetGrid(data=df, col='day', row='sex' )
grilla.map(plt.hist, 'tip')
```
