# Actividad: El Negocio de Internet

En la provincia de Soria, el negocio de acceso a Internet se reparte entre dos operadores, Timofónica y Robafone y dos únicas marcas de routers, Xisco y Nuaweii. En Soria, la cuota de mercado de Timofónica es del 60% y de Robafone el resto. El 70% de los usuarios dispone de router Xisco y el 30% de ambas marcas. Además se sabe que la probabilidad de corte de acceso es 0.1 para usuarios de Timofónica, 0.15 para Robafone y 0.05 para routers Xisco.

1. Crear un script en Python ordenado, documentado y reproducible, que defina las probabilidades proporcionadas en el enunciado (usar un diccionario).
2. Implementar en el script las expresiones correspondiente al Teorema de Bayes para responder las siguientes preguntas:
    - ¿Cuál es la probabilidad de que a un usuario se le corte el Internet?,
    - Si se sabe que un usuario tiene la línea cortada, ¿cuál es la probabilidad de que tenga router Xisco en casa?,
    - ¿Cuál es la probabilidad de que se produzca un corte a un usuario que no tiene un router Xisco?
3. Interpretar los resultados obtenidos.

```python
# Definición de probabilidades y cuotas de mercado
probabilidades = {
    'Timofonica': {
        'cuota': 0.6,
        'corte': 0.1
    },
    'Robafone': {
        'cuota': 0.4,
        'corte': 0.15
    },
    'Xisco': {
        'cuota': 0.7,
        'corte': 0.05
    },
    'Nuaweii': {
        'cuota': 0.3
    }
}

# 1. Probabilidad de que a un usuario se le corte el Internet
# P(Corte) = P(Corte|Timofonica)*P(Timofonica) + P(Corte|Robafone)*P(Robafone)
p_corte = (probabilidades['Timofonica']['corte'] * probabilidades['Timofonica']['cuota'] +
           probabilidades['Robafone']['corte'] * probabilidades['Robafone']['cuota'])
print(f"Probabilidad de corte de Internet (P(Corte)): {p_corte:.3f}")

# 2. Probabilidad de que tenga router Xisco dado que se le cortó el Internet
# P(Xisco|Corte) = [P(Corte|Xisco)*P(Xisco)] / P(Corte)
p_corte_xisco = probabilidades['Xisco']['corte'] * probabilidades['Xisco']['cuota']
p_xisco_dado_corte = p_corte_xisco / p_corte
print(f"Probabilidad de tener router Xisco dado corte (P(Xisco|Corte)): {p_xisco_dado_corte:.3f}")

# 3. Probabilidad de corte en usuarios que NO tienen router Xisco
# P(Corte|no Xisco) = P(Corte and no Xisco) / P(no Xisco)
# P(no Xisco) = 1 - P(Xisco) = 0.3
# Suponemos que la probabilidad de corte para no Xisco es la misma que la media ponderada de los operadores para ese grupo
p_no_xisco = 1 - probabilidades['Xisco']['cuota']

# No se da la probabilidad de corte para Nuaweii, así que asumimos que es la media ponderada de los operadores para ese grupo
# P(Corte) = P(Corte|Timofónica)·P(Cuota|Timofónica) + P(Corte|Robafone)·P(Cuota|Robafone) * p_no_xisco

p_corte_no_xisco = (probabilidades['Timofonica']['corte'] * probabilidades['Timofonica']['cuota'] +
           probabilidades['Robafone']['corte'] * probabilidades['Robafone']['cuota']) * p_no_xisco

# P(Corte|no Xisco) = P(Corte and no Xisco) / P(no Xisco)
p_corte_dado_no_xisco = p_corte_xisco / p_no_xisco
print(f"Probabilidad de corte en usuarios sin router Xisco (P(Corte|no Xisco)): {p_corte_dado_no_xisco:.3f}")

# Interpretación de resultados
print("\nInterpretación:")
print("1. La probabilidad de que a un usuario se le corte el Internet es {:.1f}%".format(p_corte*100))
print("2. Si se sabe que un usuario tiene la línea cortada, la probabilidad de que tenga router Xisco es {:.1f}%".format(p_xisco_dado_corte*100))
print("3. La probabilidad de corte en usuarios que no tienen router Xisco es {:.1f}%".format(p_corte_dado_no_xisco*100))
```
