# Guía: Caso Mejora Rendimiento Estudiantes

Un profesor de matemáticas, observó que en un colegio donde realizaba clases, los estudiantes se motivaban de mejor forma a revisar los contenidos cuando éstos eran presentados en un formato multimedial, como por ejemplo un video en youtube. El profesor se hizo la siguiente pregunta: ¿mejoraría el rendimiento académico de los estudiantes si incorporo en el material cápsulas de video en youtube?

Orgulloso de su idea, el profesor se presentó ante el consejo académico del colegio y planteó que la incorporación de cápsulas de video en youtube con los contenidos favorecería el rendimiento académico de los estudiantes, por lo cua solicitó financiamiento. El director, un poco escéptico, plantea que sin resultados concretos no estaría dispuesto a pedirle al sostenedor que haga una inversión sin tener seguridad absoluta de que eso daría un buen resultado.

Acuerdan, por lo tanto, realizar un piloto para ver si efectivamente las cápsulas lograban mejorar el rendimiento de los estudiantes. El piloto consistió en tomar dos colegios de la corporación educacional, de cursos y rendimientos similares, y en uno de ellos aplicar la nueva metodología basada en material en formato cápsulas de youtube, de forma de comparar los resultado y apreciar si hay mejoras significativas. Entonces, 

- Colegio Graham Parker (GP), se aplicó la nueva metodoogía
- Colegio Marteen Sheen (MS), no se aplicó.

Ambos colegios han tenido históricamente desempeños similares en las notas de sus alumnos, por lo tanto, se considera razonable tomarlos para el experimento.

Finalmente, se llevó a cabo el piloto y se registraron los resultados de ambos grupos. En una siguiente reunión, los resultados fueron presentados al consejo académico para que se tomara la decisión definitiva.
